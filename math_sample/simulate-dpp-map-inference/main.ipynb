# Simulate DPP MAP Inference (Greedy)

このノートブックでは、Determinantal Point Process (DPP) の **greedy MAP 推論**を2次元点集合上でシミュレーションし、
選択ステップをアニメーションとして可視化します。

- 点群をランダム生成
- RBF カーネルから L 行列を構築
- $\log\det(L_S)$ を目的に greedy で要素を順次選択
- 各ステップで「候補スコア」と「選択済み点」を描画
- GIF/MP4 として保存

```python
from dpp.map_simulation import run_simulation
```

```python
selected, trace, saved_path = run_simulation(
    seed=42,
    n_points=24,
    sigma=0.17,
    max_k=9,
    stop_when_nonpositive=False,
    output_dir="artifacts",
    output_stem="dpp_map_greedy",
)

print(f"Selected indices ({len(selected)}): {selected}")
print(f"Saved GIF: {saved_path}")
```
